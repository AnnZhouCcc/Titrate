```python
import numpy as np
import plotly.graph_objs as go
from collections import deque

S, S0 = 10, 10
d, d0 = 1, 2
simuLen = 500
k = 0.02
C = [10 for _ in range (100)]
# C.extend ([5 for _ in range (int (.8/k))])
# C.extend ([2.5 for _ in range (int (.8/k))])
# C.extend ([1.25 for _ in range (int (.8/k))])
C.extend ([0.625 for _ in range (400)])
queue = deque ()
Splot, dplot = [], []
for step in range (simuLen):
    S -= k * (d - d0)
    S = max (0, S)
    # queue.extend ([step for _ in range (S)])
    # d = 0
    # for _ in range (min (C, len (queue))):
    #     d = step - queue.popleft ()
    d += (S - C[step]) / C[step]
    d = max (0, d)
    Splot.append(S)
    dplot.append(d)
fig = go.Figure (data=[go.Scatter(x=list(range(len(Splot))), y=Splot, name='S')])
fig.show ()
fig = go.Figure (data=[go.Scatter(x=list(range(len(dplot))), y=dplot, name='d')])
fig.show ()
```
